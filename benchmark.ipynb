```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tools import closest_factors
```

```python
df = pd.read_csv("./benchmarks/move_sizes.csv")
names = df["puzzle_name"].unique()
rows, cols = closest_factors(len(names))
fig, subplots = plt.subplots(rows, cols, squeeze=False, figsize=(cols * 6, rows * 4))
coords_to_idx = lambda x, y: x * cols + y
width = 0.35

print(rows, cols)

for xi in range(rows):
    for yi in range(cols):
        subplot = subplots[xi, yi]
        name = names[coords_to_idx(xi, yi)]
        puzzle_df = df[df["puzzle_name"] == name]
        x = np.array(range(len(puzzle_df["move_size"])))
        subplot.bar(x - (width / 2), puzzle_df["prep_time"], label="Prep time", width=width)
        subplot.bar(x + (width / 2), puzzle_df["solve_time"], label="Solve time", width=width)
        subplot.set_xticks(x, puzzle_df["move_size"])
        subplot.set_xlabel("Move size")
        subplot.set_ylabel("Time (s)")
        subplot.set_title(name)
        subplot.legend()

fig.tight_layout()
fig.show()
```
