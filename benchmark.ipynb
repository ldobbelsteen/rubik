```python
import pandas as pd
from matplotlib import pyplot as plt
from tools import closest_factors
```

```python
# Visualize move size results benchmark.
df = pd.read_csv("./benchmarks/move_sizes.csv")
names = df["puzzle_name"].unique()
rows, cols = closest_factors(len(names))
fig, subplots = plt.subplots(rows, cols, squeeze=False)
coords_to_idx = lambda x, y: x * cols + y

for x in range(rows):
    for y in range(cols):
        subplot = subplots[x, y]
        name = names[coords_to_idx(x, y)]
        puzzle_df = df[df["puzzle_name"] == name]
        subplot.bar(puzzle_df["move_size"], puzzle_df["solve_time"], label=name)
        subplot.set_xticks(puzzle_df["move_size"])
        subplot.set_xlabel("Move size")
        subplot.set_ylabel("Solve time (s)")

fig.tight_layout()
fig.show()
```
